```python
import pandas as pd
from faker import Faker
import numpy as np
import random
from datetime import datetime, timedelta
```

```python
book_path = 'Books_df.csv'

data_book = pd.read_csv(book_path)

data_book.head(20)
```

# Sorted columns to be used

```python
book_sorted = data_book[['Title','Author','Main Genre']]
book_sorted.head(20)
```

```python
book_sorted = book_sorted.rename(columns = {'Title': 'title', 'Author': 'author','Main Genre': 'category_name'})
```

```python
book_sorted['book_id'] = range(1, len(book_sorted) + 1)
book_sorted.head()
```

```python
book_sorted = book_sorted[['book_id','title','author','category_name']]
book_sorted.head(20)
```

# Category Table

```python
df_category = pd.DataFrame(book_sorted['category'].unique(), columns = ['category_name'])
df_category.head()
```

```python
df_category['category_id'] = range(1, len(df_category) + 1)
df_category.head()
```

```python
df_category = df_category[['category_id', 'category_name']]
df_category.head()
```

category table done

# Book Table

```python
df_book = pd.merge(book_sorted, df_category, how = 'inner', on = 'category_name')
df_book.head()
```

```python
df_book = df_book[['book_id','title','author','category_id']]
df_book.head()
```

book table done

# Library Table

```python

fake = Faker('id_ID')

num_libraries = 30
libraries = []

for library_id in range(1, num_libraries + 1):
    library_name = f"Perpustakaan {fake.city()}"
    if np.random.rand() > 0.7:  
        library_name = f"Perpustakaan Umum {fake.street_name()}"
    
    libraries.append({
        'library_id': library_id,
        'name': library_name,
        'address': fake.address().replace("\n", ", "),
        'phone': fake.phone_number()
    })


df_libraries = pd.DataFrame(libraries)

df_libraries.head(10)
```

# Library_book Table

```python
num_data_lib_book = 2000
lib_book = []

for books in range(1, num_data_lib_book + 1):
    lib_book.append({
        'library_id': random.choice(df_libraries['library_id']),
        'book_id': random.choice(df_book['book_id']),
        'quantity': fake.random_int(1, 10, 1)
    })


df_library_book = pd.DataFrame(lib_book)

df_library_book.head(10)
```

# User Table

```python
num_users = 500  
start_date = datetime.now() - timedelta(days=365) 

users = []
for user_id in range(1, num_users + 1):

    fullname = fake.name()
    email = f"{fullname.split()[0].lower()}.{fake.random_number(digits=3)}@{fake.domain_name()}"
    
    created_at = fake.date_time_between(start_date=start_date, end_date="now")
    
    users.append({
        'user_id': user_id,
        'email': email,
        'password': fake.password(length=12),  
        'fullname': fullname,
        'created_at': created_at
    })
    
df_users = pd.DataFrame(users)
df_users['created_at'] = pd.to_datetime(df_users['created_at'])

df_users.head(10)
```

df_users done

# Loan Tables

```python
fake = Faker('id_ID')

num_loans = 3000
start_date = datetime.now() - timedelta(days=365) 

user_created_map = dict(zip(df_users['user_id'], df_users['created_at']))

loans = []
for loan_id in range(1, num_loans + 1):

    user_id = random.choice(df_users['user_id'])
    library_id = random.choice(df_libraries['library_id'])
    book_id = random.choice(df_book['book_id'])
    
    user_created = user_created_map[user_id]
    
    loan_date = fake.date_time_between(
        start_date=max(user_created, start_date),
        end_date='now'
    )
    
    due_date = loan_date + timedelta(days=14)
    
    status = random.choices(
        ['Active', 'Returned', 'Cancelled'],
        weights=[0.3, 0.65, 0.05],  
        k=1
    )[0]
    
    if status == 'Returned':
        if random.random() < 0.7:
            return_date = fake.date_time_between(
                start_date=loan_date, 
                end_date=due_date
            )
        else:
            return_date = fake.date_time_between(
                start_date=due_date,
                end_date=due_date + timedelta(days=random.randint(1, 60))
            )
    elif status == 'Cancelled':

        return_date = fake.date_time_between(
            start_date=loan_date,
            end_date=loan_date + timedelta(days=1))
        
    else:  
        return_date = None
        
        if random.random() < 0.7:  
            loan_date = fake.date_time_between(
                start_date=datetime.now() - timedelta(days=13),
                end_date='now'
            )
            due_date = loan_date + timedelta(days=14)

    loans.append({
        'loan_id': loan_id,
        'status': status,
        'user_id': user_id,
        'library_id': library_id,
        'book_id': book_id,
        'loan_date': loan_date,
        'due_date': due_date,
        'return_date': return_date
    })

df_loan = pd.DataFrame(loans)

df_loan = df_loan[['loan_id', 'status', 'user_id', 'library_id', 'book_id', 
                     'loan_date', 'due_date', 'return_date']]

df_loan.head(10)
```

df_loans done

# Holds Table

```python
fake = Faker('id_ID')


num_holds = 1000  
start_date = datetime.now() - timedelta(days=365) 

holds = []
for hold_id in range(1, num_holds + 1):

    user_id = random.choice(df_users['user_id'])
    library_id = random.choice(df_libraries['library_id'])
    book_id = random.choice(df_book['book_id'])
    
    hold_date = fake.date_time_between(start_date=start_date, end_date='now')
    
    expiration_date = hold_date + timedelta(days=7)
    
    current_time = datetime.now()
    if expiration_date > current_time:
        status = random.choices(
            ['Active', 'Cancelled'], 
            weights=[0.85, 0.15], 
            k=1
        )[0]
    else:
        status = random.choices(
            ['Expired', 'Fulfilled', 'Cancelled'],
            weights=[0.7, 0.25, 0.05],  
            k=1
        )[0]
    
    holds.append({
        'hold_id': hold_id,
        'user_id': user_id,
        'library_id': library_id,
        'book_id': book_id,
        'hold_date': hold_date,
        'expiration_date': expiration_date,
        'status': status
    })

df_holds = pd.DataFrame(holds)

df_holds = df_holds[['hold_id', 'status', 'library_id', 'user_id', 'book_id', 
                      'hold_date', 'expiration_date']]

df_holds.head(10)
```

df_holds done

# Exporting df created to CSV

```python
df_book.to_csv('df_book.csv', index=False)
df_category.to_csv('df_categories.csv', index=False)
df_libraries.to_csv('df_libraries.csv', index=False)
df_library_book.to_csv('df_library_book.csv', index=False)
df_loan.to_csv('df_loan.csv', index=False)
df_holds.to_csv('df_holds.csv', index=False)
df_users.to_csv('df_users.csv', index=False)
```
